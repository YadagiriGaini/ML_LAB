```python
smallDF=idf.sample(5000)
smallDF
```

```python
y=smallDF['label']
x=smallDF.drop(columns=['label'])
y_test=idf.iloc[5000:,0]
x_test=idf.iloc[5000:,1:]
```

```python
# from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
KNN.fit(x,y)
```

```python
predzero=KNN.predict(x_test)
```

```python
df_test=pd.read_csv('Week7_test.csv')
df_test.head()
# x_test=df_test.drop(columns=['label'])
pred=KNN.predict(df_test)
pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
acc=accuracy_score(y_test,predzero)
acc
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,predzero)
# print(cm)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Reds')  # You can change the color map
py.show()
```

```python
cr=classification_report(y_test,predzero)
print(cr)
```
