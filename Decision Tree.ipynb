## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
```

```python
df.fillna({'Embarked':'S'},inplace=True)
df['Embarked']
```

```python
df['Embarked'].isnull().sum()
```

```python
#Impute the missing values in Fare column with median value
```

```python
df.fillna(df['Fare'].median(),inplace=True)
df['Fare']
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
```

```python
mean_age = df['Age'].mean()  #finding mean of age column
std_age = df['Age'].std() #finding standard deviation of the column

num_missing = df['Age'].isna().sum()  # Count missing values
random_ages = np.random.normal(loc=mean_age, scale=std_age, size=num_missing)

# Step 3: Fill missing values
df.loc[df['Age'].isna(), 'Age'] = random_ages

df['Age']
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
```

```python
df['FamilySize']=df['Parch']+df['SibSp']+1 #adding new feature
df.head()
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
```

```python
# this method is called binning 

def family(size):
    if (size==1):
     return True
    else: 
     return False

df['IsAlone']=df['FamilySize'].apply(family)

df.head()
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python
def cabin(value):
    if (pd.isna(value)):
        return False
    else:
        return True
df['Has_Cabin']=df['Cabin'].apply(cabin)

df.head()
```

```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
df.head()
```

## Model Building

```python
#Prepare X and y
```

```python
y=df['Survived']
X=df.drop(columns=['Survived'])
X
```

```python
# Label encoding of categorical features
```

```python
categorical=df.select_dtypes(include=['object','bool']).columns

categorical
```

```python
from sklearn.preprocessing import LabelEncoder

LAB = LabelEncoder()

for cols in categorical:
    df[cols]=LAB.fit_transform(df[cols])

df.head()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics
```

```python
model = DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
#Print the tree

# Importing required packages for visualization
```

```python
plt.figure(figsize=(12, 8))
tree.plot_tree(model, feature_names=X.columns, class_names=["Not Survived","Survived"], filled=True)
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score

# Importing classification report and confusion matrix from sklearn metrics

# Making predictions

# Printing classification report
```

```python
y_pred = model.predict(x_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
precision = metrics.precision_score(y_test, y_pred)  # Positive class default is 1
recall = metrics.recall_score(y_test, y_pred)
f1 = metrics.f1_score(y_test, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```
